# Comparison of f- and t-domain half-space solutions

Comparing `kernel.halfspace` with `dipole` for a half-space and a fullspace-solution, where `dipole` internally uses `kernel.fullspace` for the fullspace solution (`xdirect=True`).

```python
import numpy as np
from matplotlib import rcParams
import matplotlib.pyplot as plt

from empymod import dipole, utils
from empymod.kernel import halfspace

# Style adjustments
%matplotlib inline
plt.style.use('ggplot')
rcParams['figure.dpi'] = 300
rcParams['savefig.dpi'] = 300
rcParams['text.usetex'] = True
rcParams['font.serif'] = 'Computer Modern Roman'
rcParams['font.family'] = 'serif'
rcParams['font.style'] = 'normal'
```

## Time Domain

```python
src = [0, 0, 100]
rec = [2000, 500, 200]
res = [2e14, 2.5]
aniso = [1, 2]
t = np.logspace(-2, 3, 301)

# Collect parameters
inpEM = {'src': src, 'rec': rec,'freqtime': t, 'verb': 0}
modHS = {'depth': 0, 'res': res, 'aniso': aniso}
modFS = {'depth': [], 'res': res[1], 'aniso': aniso[1]}
inpAN = {'zsrc': src[2], 'xco': rec[0], 'yco': rec[1], 'zrec': rec[2], 'res': res[1], 'aniso': aniso[1], 'freqtime': t}
```

```python
def plot_t(EM, HS, title, i):
    plt.figure(title)
    plt.subplot(i)
    plt.semilogx(t, EM)
    plt.semilogx(t, HS, '--')  
    
i = 330
for ab in [11, 12, 13, 21, 22, 23, 31, 32, 33]:
    i += 1
    
    ### Halfspace
    
    # Impulse
    EM = dipole(**inpEM, **modHS, ab=ab, signal=0)
    HS = halfspace(**inpAN, ab=ab, signal=0)
    plot_t(EM, HS, 'Impulse HS', i)
    
    # Switch-on
    EM = dipole(**inpEM, **modHS, ab=ab, signal=1)
    HS = halfspace(**inpAN, ab=ab, signal=1)
    plot_t(EM, HS, 'Switch-on HS', i)
    
    # Switch-on
    EM = dipole(**inpEM, **modHS, ab=ab, signal=-1)
    HS = halfspace(**inpAN, ab=ab, signal=-1)
    plot_t(EM, HS, 'Switch-off HS', i)

    ###  Fullspace
    
    # Impulse
    EM = dipole(**inpEM, **modFS, ab=ab, signal=0)
    HS = halfspace(**inpAN, ab=ab, signal=0, output='fs')
    plot_t(EM, HS, 'Impulse FS', i)
    
    # Switch-on
    EM = dipole(**inpEM, **modFS, ab=ab, signal=1)
    HS = halfspace(**inpAN, ab=ab, signal=1, output='fs')
    plot_t(EM, HS, 'Switch-on FS', i)
    
    # Switch-on
    EM = dipole(**inpEM, **modFS, ab=ab, signal=-1)
    HS = halfspace(**inpAN, ab=ab, signal=-1, output='fs')
    plot_t(EM, HS, 'Switch-off FS', i)

plt.figure('Impulse HS')
plt.suptitle('Impulse HS')
plt.figure('Switch-on HS')
plt.suptitle('Switch-on HS')
plt.figure('Switch-off HS')
plt.suptitle('Switch-off HS')
plt.figure('Impulse FS')
plt.suptitle('Impulse FS')
plt.figure('Switch-on FS')
plt.suptitle('Switch-on FS')
plt.figure('Switch-off FS')
plt.suptitle('Switch-off FS')

plt.show()
```

## Frequency domain

```python
inpEM['freqtime'] = 1/t
inpAN['freqtime'] = 1/t

def plot_f(EM, HS, title, i):
    plt.figure(title)
    plt.subplot(i)
    plt.semilogx(1/t, EM.real)
    plt.semilogx(1/t, HS.real, '--')
    plt.semilogx(1/t, EM.imag)
    plt.semilogx(1/t, HS.imag, '--')

i = 330
for ab in [11, 12, 13, 21, 22, 23, 31, 32, 33]:
    i += 1
    
    ### Halfspace
    EM = dipole(**inpEM, **modHS, ab=ab)
    HS = halfspace(**inpAN, ab=ab)
    plot_f(EM, HS, 'Frequency HS', i)

    ###  "Fullspace"
    EM = dipole(**inpEM, **modFS, ab=ab)
    HS = halfspace(**inpAN, ab=ab, output='fs')
    plot_f(EM, HS, 'Frequency FS', i)
    

plt.figure('Frequency HS')
plt.suptitle('Frequency HS')
plt.figure('Frequency FS')
plt.suptitle('Frequency FS')
plt.show()
```
